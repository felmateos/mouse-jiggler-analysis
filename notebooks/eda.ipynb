```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/exp/random.csv')
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=df["x"], 
    y=df["y"], 
    fill=True,   # deixa preenchido
    cmap="viridis",  # esquema de cores
    thresh=0,    # mostra áreas de baixa densidade
    levels=100   # mais suavização
)

plt.title("Heatmap das posições do mouse")
plt.gca().invert_yaxis()  # opcional: inverter Y (sistema de coordenadas da tela)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hexbin(df["x"], df["y"], gridsize=50, cmap="inferno")
plt.colorbar(label="Frequência")
plt.title("Heatmap das posições do mouse (Hexbin)")
plt.gca().invert_yaxis()
plt.show()
```
